```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.preprocessing import StandardScaler, LabelEncoder
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
df=pd.read_csv('tox21/tox21_dense_train.csv')
```

```python
dense_train=df
dense_test=pd.read_csv('tox21/tox21_dense_test.csv')
labels_test=pd.read_csv('tox21/tox21_labels_test.csv')
labels_train=pd.read_csv('tox21/tox21_labels_train.csv')
```

```python
train_y=labels_train['NR.AhR'].dropna()
train_x=dense_train.loc[train_y.index].drop(columns=['Unnamed: 0'])
train_x=StandardScaler().fit_transform(train_x)
```

```python
encoder = LabelEncoder()
encoder.fit(train_y)
encoded_train = encoder.transform(train_y)
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=801))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_x[:100], encoded_train[:100], epochs=10, batch_size=10)
```
